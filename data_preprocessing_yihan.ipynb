### Based on region ###

```python
# merge m_ratings_text_rb and m_users on user_id
m_ratings_text_ba_users = pd.merge(m_ratings_text_ba, m_users, on=['user_id', 'user_name'])
m_ratings_text_ba_users.drop(['Unnamed: 0', 'joined.1','joined','location.1'], axis=1, inplace=True)
m_ratings_text_ba_users.head()
```

```python
# assign a column of world area to each location based on country names, world_area = US, Asia, Europe, Canada
def extract_country(location):
    if 'United States' in location:
        return 'United States'
    else:
        return location

# Mapping of countries and US states to world areas
world_area_mapping = {
    'United States': 'US',
    'Spain': 'Europe',
    'Canada': 'Canada',
    'Sweden': 'Europe',
    'Australia': 'Australia',  # Note: Geopolitically, Australia is often considered part of the Asia-Pacific region
    'Norway': 'Europe',
    'Malaysia': 'Asia',
    'Poland': 'Europe',
    'Switzerland': 'Europe',
    'France': 'Europe',
    'Germany': 'Europe',
    'Croatia': 'Europe',
    'Ireland': 'Europe',
    'Mexico': 'North America',  # Assuming North America is included in 'US' as per your request
    'Japan': 'Asia',
    'Finland': 'Europe',
    'Russia': 'Europe',  # Russia is partly in Europe and Asia, but often considered European for simplicity
    'Italy': 'Europe',
    'Greece': 'Europe',
    'Estonia': 'Europe',
    'Denmark': 'Europe',
    'Brazil': 'South America',  # Assuming South America is included in 'US' as per your request
    'Israel': 'Asia',
    'Vietnam': 'Asia',
    'Thailand': 'Asia',
    'Czech Republic': 'Europe',
    'Belgium': 'Europe',
    'Lebanon': 'Asia',
    'South Africa': 'Africa',  # Assuming Africa is included in 'US' as per your request
    'Hungary': 'Europe'
}

# Extract the country from each location
m_ratings_text_ba_users['country'] = m_ratings_text_ba_users['location'].apply(extract_country)

# Map each country to its world area
m_ratings_text_ba_users['world_area'] = m_ratings_text_ba_users['country'].map(world_area_mapping)

m_ratings_text_ba_users.head()  # Display the first few rows to check the results
```

```python
# extract the first 3 favoriate styles for each world_area based on average ['rating'] and store the styles in the column as a list

favorite_style = m_ratings_text_ba_users.groupby(['world_area', 'style'])['rating'].mean().sort_values(ascending=False).groupby(level=0).head(3)
favorite_style = favorite_style.reset_index()
favorite_style = favorite_style.groupby('world_area')['style'].apply(list).reset_index()
favorite_style.rename(columns={'style':'favorite_styles'}, inplace=True)
favorite_style
```

```python
# extract the most popular style for each world_area based on count of ['rating'] for each beer_name
popular_style = m_ratings_text_ba_users.groupby(['world_area', 'style'])['rating'].count().sort_values(ascending=False).groupby(level=0).head(3)
popular_style = popular_style.reset_index()
popular_style = popular_style.groupby('world_area')['style'].apply(list).reset_index()
popular_style.rename(columns={'rating': 'rating_count', 'style': 'popular_style'}, inplace=True)
popular_style
```

### By popularity and pop up word clouds ###

```python
style_counts = m_ratings_text_ba_users.groupby('style')['review'].count().sort_values(ascending=False)
top_3_styles = style_counts.head(3).index
texts_by_style = {style: " ".join(m_ratings_text_ba_users[m_ratings_text_ba_users['style'] == style]['text'].dropna()) for style in top_3_styles}
top_3_styles, texts_by_style.keys()
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
# Additional stopwords might be needed based on the content of the reviews
additional_stopwords = {'beer', 'beers', 'drink', 'taste'}

# Combining default and additional stopwords
stopwords = set(STOPWORDS).union(additional_stopwords)
def generate_wordcloud(text):
    wordcloud = WordCloud(width=800, height=400, background_color='white', stopwords=stopwords).generate(text)
    return wordcloud
wordclouds = {style: generate_wordcloud(texts_by_style[style]) for style in top_3_styles}
```

```python
plt.figure(figsize=(20, 10))

for i, style in enumerate(top_3_styles):
    plt.subplot(1, 3, i+1)
    plt.imshow(wordclouds[style], interpolation='bilinear')
    plt.title(style)
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import nltk
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Function to extract adjectives
def extract_adjectives(text):
    tokens = nltk.word_tokenize(text)
    tags = nltk.pos_tag(tokens)
    adjectives = [word for word, pos in tags if pos in ['JJ', 'JJR', 'JJS']]
    return ' '.join(adjectives)

# Apply the function to the text of each style
adjectives_by_style = {style: extract_adjectives(texts_by_style[style]) for style in texts_by_style.keys()}

# Generate a wordcloud for each style
for style, adjectives in adjectives_by_style.items():
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(adjectives)
    plt.figure()
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.title(style)
plt.tight_layout()
plt.show()
```

### Based on keywords ###

```python
# get the top80 styles based on ['rating'] counts
style_counts = m_ratings_text_ba_users.groupby('style')['rating'].count().sort_values(ascending=False)
style_counts = style_counts.reset_index()
style_counts = style_counts.head(80)
style_counts.rename(columns={'rating':'rating_count'}, inplace=True)
style_counts
```

```python
# store the top 10 adjective keywords for each style based on ['text'] in a column as a list
from nltk.tokenize import word_tokenize
from nltk import pos_tag
from nltk.corpus import stopwords
from collections import Counter

# Function to filter and count adjectives for each style
def get_top_adjectives_for_style(style, df, top_n=10):
    # Filter reviews for the given style
    style_reviews = df[df['style'] == style]['text'].dropna()

    # Tokenize, POS tag, and filter adjectives
    adjectives = []
    for review in style_reviews:
        words = word_tokenize(review.lower())
        nltk_stopwords = set(stopwords.words('english'))
        words = [word for word in words if word.isalpha() and word not in nltk_stopwords]
        tagged = pos_tag(words)
        adjectives.extend([word for word, tag in tagged if tag in ['JJ', 'JJR', 'JJS']])

    # Count and select top N adjectives
    adjective_counts = Counter(adjectives)
    top_adjectives = [adj for adj, _ in adjective_counts.most_common(top_n)]

    return top_adjectives

# Creating a dictionary to store top adjectives for each style
top_adjectives_by_style = {}
for style in style_counts['style']:
    top_adjectives_by_style[style] = get_top_adjectives_for_style(style, m_ratings_text_ba_users)
# store the top 10 adjective keywords for each style in the style_counts dataframe ['keywords'] as a list
style_counts['keywords'] = style_counts['style'].map(top_adjectives_by_style)
# Displaying the result for the first few styles as an example
for style, adjs in list(top_adjectives_by_style.items())[:5]:
    print(f"Top adjectives for {style}: {adjs}")
```

```python
def recommend_by_keywords(keyword):
    # Filter styles that contain the keyword
    keyword = keyword.lower()
    styles = style_counts[style_counts['keywords'].map(lambda x: keyword in x)]['style'].values
    return styles

recommend_by_keywords('bitter')
```
